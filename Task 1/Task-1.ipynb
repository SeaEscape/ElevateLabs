```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")

original_head =df.head() #First 4 entries 
```

```python
original_tail= df.tail() #Last 4 entries
```

```python
#Capture Initial rows count for duplicate calculation
initial_rows = len(df)
```

```python
# 1.Identify and handle missing values using .isnull() in Python or filters in Excel
print("Missing values before cleaning :")
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#fill the space with Unknown values
df['director']=df['director'].fillna('Unknown')
df['cast']=df['cast'].fillna('Unknown')
df['country']=df['country'].fillna('Unknown')
df['date_added']=df['date_added'].fillna('Unknown')
df['rating']=df['rating'].fillna('Unknown')
df['duration']=df['duration'].fillna('Unknown')
```

```python
#2. Remove duplicate rows using .drop_duplicates() or Excel’s “Remove Duplicates”.
df=df.drop_duplicates()

duplicates_removed = initial_rows -len(df)
df
```

```python
#3.Standardize text values like gender, country names, etc.
def standardize_text(text):
    if pd.isna(text) or text == 'Unknown':
        return text
    text = ' '.join(text.strip().split())
    return text.title()

text_columns = ['title','director','cast','country','rating','listed_in','description']
for col in text_columns:
    df[col] = df[col].apply(standardize_text)

#Standardize country names
df['country']=df['country'].replace({
    'United States':'USA',
    'United Kingdom' : 'UK',
    'South Africa' : 'SA'
})
```

```python
#4. Convert data formats to dd-mm-yyyy
def standardize_date(date_str):
    if date_str ==  'Unknown':
        return pd.NA
    try:
        cleaned_date=" ".join(date_str.strip().split())
        parsed_date = pd.to_datetime(cleaned_date,errors='coerce')
        return parsed_date.strftime('%d-%m-%Y') if not pd.isna(parsed_date) else pd.NA
    except:
        return pd.NA
```

```python
#5.Rename column headers to be clean and uniform (e.g., lowercase, no spaces).
df = df.rename(columns={
    'type':'Content_types',
    'listed_in':'Genres'
})
df.columns = [col.lower().replace(" ","_") for col in df.columns]
```

```python
#6.Check and fix data types (e.g., age should be int, date as datetime).
df['release_year'] = pd.to_numeric(df['release_year'],errors='coerce',downcast='integer')

def parse_duration(duration):
    if pd.isna(duration):
        return duration
    parts = duration.split()
    try:
        return int(parts[0]) if parts[0].isdigit() else duration
    except:
        return duration

df['duration']= df['duration'].apply(parse_duration)
```

```python
# Capture cleaned head, tail, and data types
cleaned_head = df.head()
cleaned_tail = df.tail()
cleaned_dtypes = df.dtypes
```

```python
#Save cleaned dataset
df.to_csv('netflix_titles_cleaned.csv', index=False)
```

```python
# Store results
results = {
    'original_head': original_head.to_dict(),
    'original_tail': original_tail.to_dict(),
    'cleaned_head': cleaned_head.to_dict(),
    'cleaned_tail': cleaned_tail.to_dict(),
    'data_types': cleaned_dtypes.to_dict()
}
```
